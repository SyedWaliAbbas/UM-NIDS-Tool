Extracting 5,000 or fewer samples for each specified class.

```python
# Function to sample up to 5000 rows from each label
def sample_labels(df, labels, label_col='label', max_samples=5000):
    sampled_df = df[df[label_col].isin(labels)].groupby(label_col).apply(lambda x: x.sample(n=min(len(x), max_samples), random_state=42))
    return sampled_df.reset_index(drop=True)
```

```python
df_cic_iot = sample_labels(df_cic_iot, cic_iot_2023_labels)
```

```python
df_unsw = sample_labels(df_unsw, unsw_labels)
df_cic_ddos = sample_labels(df_cic_ddos, cic_ddos_labels)
df_cic_17 = sample_labels(df_cic_17, cic_ids_2017_labels)
```

Combining all to form a unified dataset

```python
# Optionally, you can concatenate all sampled datasets into one unified dataset
df_unified_sampled = pd.concat([df_cic_17, df_cic_iot, df_unsw, df_cic_ddos])
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score

# Assuming you have already created 'df_unified_sampled' from previous steps

# Step 1: Encode the labels and prepare the feature matrix
label_encoder = LabelEncoder()

# Assuming 'label' column is the target
df_unified_sampled['label_encoded'] = label_encoder.fit_transform(df_unified_sampled['label'])

# X will contain all features except the label columns
X = df_unified_sampled.drop(['label', 'label_encoded'], axis=1)

# Y will be the encoded labels
y = df_unified_sampled['label_encoded']

# Step 2: Split into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Train a Random Forest Classifier
rf_clf = RandomForestClassifier(random_state=42,n_jobs=-1)
rf_clf.fit(X_train, y_train)

# Make predictions
y_pred_rf = rf_clf.predict(X_test)

# Evaluate the classifier
print(f"Random Forest Accuracy: {accuracy_score(y_test, y_pred_rf):.4f}")
print(f"Classification Report for Random Forest:\n{classification_report(y_test, y_pred_rf, target_names=label_encoder.classes_)}")
```

```python
from sklearn.tree import DecisionTreeClassifier

# Train a Decision Tree Classifier
dt_clf = DecisionTreeClassifier(random_state=42)
dt_clf.fit(X_train, y_train)

# Make predictions
y_pred_dt = dt_clf.predict(X_test)

# Evaluate the classifier
print(f"Decision Tree Accuracy: {accuracy_score(y_test, y_pred_dt):.4f}")
print(f"Classification Report for Decision Tree:\n{classification_report(y_test, y_pred_dt, target_names=label_encoder.classes_)}")
```

```python
from sklearn.neural_network import MLPClassifier

# Train a Neural Network (MLP) Classifier
mlp_clf = MLPClassifier(random_state=42, max_iter=300)
mlp_clf.fit(X_train, y_train)

# Make predictions
y_pred_mlp = mlp_clf.predict(X_test)

# Evaluate the classifier
print(f"MLP Neural Network Accuracy: {accuracy_score(y_test, y_pred_mlp):.4f}")
print(f"Classification Report for MLP Neural Network:\n{classification_report(y_test, y_pred_mlp, target_names=label_encoder.classes_)}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score

# Initialize K-Nearest Neighbors model
knn = KNeighborsClassifier()

# Train the model
knn.fit(X_train, y_train)

# Make predictions
y_pred_knn = knn.predict(X_test)

# Evaluate the model
print("K-Nearest Neighbors Results:")
print(classification_report(y_test, y_pred_knn))
print("Accuracy:", accuracy_score(y_test, y_pred_knn))
```
